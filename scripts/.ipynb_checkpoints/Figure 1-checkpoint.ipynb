```python
import numpy as np
import xarray as xr

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import dates as mdates
import matplotlib.patches as mpatches
import matplotlib.ticker as mticker

import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io import shapereader

import geopandas as gpd
import regionmask
```

```python
def mask_data(arr_shape, states, lon, lat, conus_shapefile_path='shp/CONUS.shp'):
    r"""Masks an input xarray dataset using the
    two character state IDs in states based on the 
    lon_name and lat_name coordinates in input_dataset and
    if they are within the given list of states. Original dataset 
    is not modified.
    
    Parameters
    ----------
    input_dataset: xarray dataset
        Dataset that you want to mask. Must have a lon and lat coordinate.
    states: list or ndarray
        List of states used to extract data subset.      
    lon_name: str
        Name of longitude coordinate in input_dataset. Default is 'lon'.
    lat_name: str
        Name of latitude coordinate in input_dataset. Default is 'lat'.
    conus_shapefile_path: str
        Path to shapefile used to extract geography information.
        
    Returns
    -------
    output_dataset: xarray dataset
        Modified input_dataset where grids not in states are set to nan.
    """
    
    ones=np.ones(arr_shape)
    
    usa = gpd.read_file(conus_shapefile_path)
    usa = usa[usa.STUSPS.isin(states)]

    state_mask = regionmask.mask_geopandas(usa, lon, lat)
    
    ma = state_mask.values
    ma[~np.isnan(ma)] = 1
    output_dataset = ones * ma
    
    return output_dataset
```

```python
regions = {'SP': ['TX', 'OK', 'KS'],
           'NP': ['NE', 'SD', 'ND'],
           'MW': ['WI', 'MI', 'OH', 'IL', 'IN', 'MN', 'IA', 'MO', 'KY'],
           'SE': ['AR', 'LA', 'TN', 'MS', 'AL', 'SC', 'NC', 'FL', 'GA'],
           'NE': ['NJ', 'PA', 'NY', 'VT', 'NH', 'CT', 'VA', 'WV', 'MD', 'DC', 'DE', 'RI', 'MA', 'ME'],}
           #'EC': []}

#ECONUS is comprised of states in all other regions
#[regions['EC'].extend(regions[x]) for x in regions.keys() if x!='EC'];


region_id = {#'EC': 'ECONUS', 
                'NP': 1, 
                'SP': 2, 
                'MW': 3, 
                'SE': 4, 
                'NE': 5}

region_names = {#'EC': 'ECONUS', 
                1: 'Northern Plains', 
                2: 'Southern Plains', 
                3: 'Midwest', 
                4: 'Southeast', 
                5: 'Northeast'}
```

```python
geog=xr.open_dataset('../data/geog_info.nc')
```

```python
#region_string_array=np.repeat(np.array(['Undefined']*1399)[np.newaxis],899,axis=0)
region_string_array=np.repeat(np.array([0]*1399)[np.newaxis],899,axis=0)

for region in regions.keys():
    print(region)
    masked_states=xr.DataArray(mask_data(geog.HGT_M[0].shape,regions[region],
                       geog.XLONG_M[0],geog.XLAT_M[0]))
    
    region_string_array=np.where(~np.isnan(masked_states),region_id[region],region_string_array)
```

```python
region_xr_array=xr.DataArray(region_string_array)
#region_xr_array.to_netcdf('data/region_mask.nc')
```

```python
region_xr_array.plot()
```

```python
# Set some projections for our data (Plate Carree)
# and output maps (Lambert Conformal)

# Data projection;
dataproj = ccrs.PlateCarree()

# Plot projection
# The look you want for the view, LambertConformal for mid-latitude view
plotproj = ccrs.LambertConformal(central_longitude=-96.,
                                 central_latitude=40.,
                                 standard_parallels=[30, 60])
```

```python
wrf_lat=geog.XLAT_M[0]
wrf_lon=geog.XLONG_M[0]

# request data for use by geopandas
resolution = '110m'
category = 'cultural'
name = 'admin_0_countries'

shpfilename = shapereader.natural_earth(resolution, category, name)
df = gpd.read_file(shpfilename)
df=df[df.TYPE=='Country']
USA=df[df.SOV_A3=='US1']

shpfilename_lakes = shapereader.natural_earth('110m', 'physical', 'lakes')
lakes_df = gpd.read_file(shpfilename_lakes)

lakes_df=lakes_df[lakes_df.name_alt=='Great Lakes']#.geometry

glakes_mask=regionmask.mask_geopandas(lakes_df.geometry,wrf_lon,wrf_lat)
usa_mask=regionmask.mask_geopandas(USA.geometry,wrf_lon,wrf_lat)
```

```python
colors=['none','red','orange','green','blue','violet']
```

```python
zip_latlons=[]
for i,val in enumerate(wrf_lat[0,:]):
    zip_latlons.append((wrf_lon[0,i],val))
for i,val in enumerate(wrf_lat[:,-1]):
    zip_latlons.append((wrf_lon[i,-1],val))
for i,val in enumerate(wrf_lat[-1][::-1]):
    zip_latlons.append((wrf_lon[-1][::-1][i],val))
for i,val in enumerate(wrf_lat[:,0][::-1]):
    zip_latlons.append((wrf_lon[:,0][::-1][i],val))
```

```python
fig=plt.figure(constrained_layout=True, figsize=(10,10))
gs=fig.add_gridspec(1,1)

ax=fig.add_subplot(gs[0],projection=plotproj)
ax.set_extent([-124., -71, 22., 53.],ccrs.PlateCarree())
ax.add_feature(cfeature.NaturalEarthFeature('physical', 'land', '50m', facecolor='blanchedalmond'))
ax.add_feature(cfeature.NaturalEarthFeature('physical', 'ocean', '50m', facecolor='lightblue'))

ax.add_feature(cfeature.COASTLINE,linewidth=0.5)
ax.add_feature(cfeature.STATES,linewidth=0.5,zorder=6)

c=ax.contourf(wrf_lon,wrf_lat,
                geog.HGT_M[0].where(~np.isnan(usa_mask.values) & np.isnan(glakes_mask.values)),
                np.arange(0,5500,500),cmap='Greys',
                transform=dataproj,zorder=5)

##########################################################################################
dashed_region=True
if dashed_region:
    polygons={}
    for i,r in enumerate(np.unique(region_xr_array)):
        if not r:
            continue
        mpl.rcParams['hatch.color']=colors[i]
        
        if r==5:
            cr=ax.contourf(wrf_lon,wrf_lat,
                           region_xr_array.where(region_xr_array==r),
                           colors=['none','none'],hatches=['','/'],transform=dataproj,zorder=6)
        else:
            cr=ax.contourf(wrf_lon,wrf_lat,
                           region_xr_array.where(region_xr_array==r),
                           colors=['none','none'],hatches=['/',''],transform=dataproj,zorder=6)


        #something weird going on with the contours on NE, need to specify different collection
        if r==5:
            contour=cr.collections[3].get_paths()[0]
        else:
            contour=cr.collections[0].get_paths()[0]
        polygons[r]=contour.to_polygons()[0]
        ax.plot(polygons[r][:,0],polygons[r][:,1],
                color=colors[i],linewidth=3,
                transform=dataproj,linestyle='--',zorder=5)
##########################################################################################


np_poly=plt.Polygon([(0,0),(0,0)],ec='red',facecolor='none',lw=3,ls='--',zorder=0,
                transform=dataproj)
sp_poly=plt.Polygon([(0,0),(0,0)],ec='orange',facecolor='none',lw=3,ls='--',zorder=0,
                transform=dataproj)
mw_poly=plt.Polygon([(0,0),(0,0)],ec='green',facecolor='none',lw=3,ls='--',zorder=0,
                transform=dataproj)
se_poly=plt.Polygon([(0,0),(0,0)],ec='blue',facecolor='none',lw=3,ls='--',zorder=0,
                transform=dataproj)
ne_poly=plt.Polygon([(0,0),(0,0)],ec='violet',facecolor='none',lw=3,ls='--',zorder=0,
                transform=dataproj)

poly=plt.Polygon(zip_latlons,ec='black',facecolor='none',lw=3,ls='--',zorder=10,
                transform=dataproj)

ax.add_patch(poly)
ax.add_patch(sp_poly)
ax.add_patch(np_poly)
ax.add_patch(mw_poly)
ax.add_patch(se_poly)
ax.add_patch(ne_poly)

ax.legend([poly,np_poly,sp_poly,mw_poly,se_poly,ne_poly],
          ['WRF Simulation Extent','Northern Plains', 'Southern Plains','Midwest',
          'Southeast','Northeast'],
          loc='lower center',bbox_to_anchor=(0.5,-0.3),ncol=3,fontsize=15)


cbar=plt.colorbar(c,orientation='vertical',aspect=30,fraction=0.021,pad=0.01)
cbar.set_label('Terrain Height (m)',fontsize=15)

ll=ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,linestyle=':',lw=1.2,zorder=11)
ll.top_labels=False
ll.right_labels=False
ll.xlocator = mticker.FixedLocator([-120,-110,-100,-90,-80,-70])


plt.show()
plt.close()
```
