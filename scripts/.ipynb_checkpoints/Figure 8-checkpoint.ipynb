```python
import pickle
import numpy as np
import xarray as xr
import pandas as pd

import dask
import dask.array as da
from dask import delayed

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import dates as mdates
import matplotlib.patheffects as PathEffects
from matplotlib.gridspec import GridSpec

import cartopy.crs as ccrs
import cartopy.feature as cfeature

from scipy.ndimage.measurements import center_of_mass
from scipy.stats import mannwhitneyu, pearsonr, levene

from glob import glob

from datetime import datetime, timedelta

import scipy.stats as stats
```

```python
def load_and_clean_data(files):
    df_list=[]
    maxes=[]
    current_water_year=0

    #loop through list of files - these should be strings using glob or something
    for f in files:
        print(f)
        
        #read csv
        temp_load=pd.read_csv(f)
        
        #change index to datetime
        #because certain times repeat for concurrent storms, this isn't recommended for index
        #temp_load=temp_load.set_index(pd.DatetimeIndex(temp_load.datetime).rename('index')).drop('datetime',axis=1)
        
        #add a column called water_year - this value ticks up by 1 for each file
        temp_load['water_year']=current_water_year

        #pull in the max labeled storm
        maxes.append(temp_load['storm_num'].max()+1)

        #for every file after the first, add the max of the previous appended dataframe to the
        #storm_num category. this is because storm_num resets in each water year. when concatenating
        #into one df, there can't be any overlap
        if df_list:
            temp_load['storm_num']=temp_load['storm_num']+df_list[-1]['storm_num'].max()+1

        #append the df into a list
        df_list.append(temp_load)

        #tick the water year up
        current_water_year+=1

    #save the maxes in case you need them later
    maxes=np.array(maxes)

    #return concatenated list into one big df + the maxes
    return pd.concat(df_list).reset_index(),maxes

def calc_distance(group):
    m_displaced = np.sqrt((4000*(group['xind'].values[-1] - group['xind'].values[0]))**2 + (4000*(group['yind'].values[-1] - group['yind'].values[0]))**2)
    speed_ms = m_displaced/(len(group)*15*60)
    return speed_ms #speed in m/s
```

```python
#year doesn't matter here, we just want month and day for 365 days
days_of_year=np.arange(datetime(1990,10,1), datetime(1991,10,1), timedelta(days=1)).astype(datetime)
```

```python
%%time

sims=['historical','future_8p5']

data={}
for sim in sims:  
    print(sim)
    files=glob('/eagle/climate_severe/bwallace_scratch/mcs_pkl_no_coords/mcs_tracks*'+sim+'*')
    files.sort()
    
    data[sim],data[sim+'_maxes']=load_and_clean_data(files)
        
    #load in our seperately calculated dataframe of storm centroids 
    centroid_path=glob('/eagle/climate_severe/bwallace_scratch/mcs_pkl_no_coords/*'+sim+'*centroid*')
    curr_centroids=pd.read_csv(centroid_path[0])
    
    precip_path=glob('/eagle/climate_severe/bwallace_scratch/mcs_pkl_no_coords/*'+sim+'*precip*')
    curr_precip=pd.read_csv(precip_path[0])
    
    #match the storm_id in the centroid dataframe against our tracks dataframe and
    #fill in the values - takes about ~2 minutes with 17,000 unique storms
    for storm_to_match in data[sim]['storm_num'].unique():
        indices=data[sim][data[sim]['storm_num']==storm_to_match].index
        centroid_indices=curr_centroids[curr_centroids['storm_id']==storm_to_match].index
        precip_indices=curr_precip[curr_precip['storm_id']==storm_to_match].index

        data[sim].loc[indices,'region_centroid']=curr_centroids.loc[centroid_indices,'region_centroid'].values
        data[sim].loc[indices,'region_wgt_centroid']=curr_centroids.loc[centroid_indices,'region_wgt_centroid'].values
        data[sim].loc[indices,'xind']=curr_centroids.loc[centroid_indices,'xind'].values
        data[sim].loc[indices,'yind']=curr_centroids.loc[centroid_indices,'yind'].values
        data[sim].loc[indices,'accum_precip']=curr_precip.loc[precip_indices,'precip_sum'].values

    #calculate precip volume & life-time precip volume
    data[sim]['pvolume']=data[sim]['area']*(data[sim]['accum_precip']*1e-6)
    data[sim]['lifetime_pvolume']=data[sim].groupby(['storm_num'])['pvolume'].transform('sum')
    
    #retrieve unique regions from dataframe
    unique_regions=data[sim]['region_centroid'].unique()
    
    data[sim]['speed'] = data[sim].groupby('storm_num').apply(calc_distance)
    data[sim]['speed'] = data[sim].loc[data[sim]['storm_num'],'speed'].values
```

```python
def scale_down_data(dset,ur,var):
    #if our input variable is duration (costant quantity per storm_id)
    if var=='duration':        
        #select our appropriate region
        if ur=='Undefined':
            temp=dset[dset['region_centroid']!=ur]
        else:
            temp=dset[dset['region_centroid']==ur]
            
        #drop duplicates storm instances since this value is constant for the life of the storm
        temp=temp.drop_duplicates(subset=['storm_num'])

        #return the average storm duration for a specific region within each water year
        return temp.set_index(pd.DatetimeIndex(temp.datetime))[var].resample('AS-OCT').mean()
    
    #if var is not duration
    else:
        #group by region again
        if ur=='Undefined':
            temp=dset[dset['region_centroid']!=ur]
        else:
            temp=dset[dset['region_centroid']==ur]

        #get all unique MCS
        all_unique_stormids=temp['storm_num'].unique()
        
        #loop through each unique MCS id
        final=[]
        for individual_stormid in all_unique_stormids:
            temp_stormid=temp[temp['storm_num']==individual_stormid]
            #append the point in the storm life where it equals its max lifetime value for the var
            final.append(temp_stormid[temp_stormid[var]==temp_stormid[var].max()])

        #concat list into a pd dataframe again
        final=pd.concat(final)

        #return the average for each water year
        return final.set_index(pd.DatetimeIndex(final.datetime))[var].resample('AS-OCT').mean()

def scale_down_data_mean(dset,ur,var):
    #if our input variable is duration or speed (constant quantity per storm_id)
    if var=='duration' or var=='speed':        
        
        #You should probably keep it this way. Only consider duration for storms that originate in each region
        temp=dset.drop_duplicates(subset=['storm_num'])

        #select our appropriate region
        if ur=='Undefined':
            #temp=dset[dset['region_centroid']!=ur]
            temp=temp[temp['region_centroid']!=ur]

        else:
            #temp=dset[dset['region_centroid']==ur]
            temp=temp[temp['region_centroid']==ur]
            
        #drop duplicates storm instances since this value is constant for the life of the storm
        #temp=temp.drop_duplicates(subset=['storm_num'])

        #return the average storm duration for a specific region within each water year
        return temp.set_index(pd.DatetimeIndex(temp.datetime))[var].resample('AS-OCT').mean()
    
    #if var is not duration
    else:
        #group by region again
        if ur=='Undefined':
            temp=dset[dset['region_centroid']!=ur]
        else:
            temp=dset[dset['region_centroid']==ur]

        #return the average for each water year
        return temp.groupby('storm_num')[var].mean().reset_index().set_index(pd.DatetimeIndex(temp.groupby('storm_num')['datetime'].first())).resample('AS-OCT').mean()[var]
```

```python
simulation_color_codes = {'historical': '#c7e9c0', 
                          'future_4p5': '#6baed6', 
                          'future_8p5': '#a50f15'}

#modify the default boxplot
meanpointprops = dict(marker='.', markeredgecolor='black',
                      markerfacecolor='black')

y_labs={
    'area':' Area $\geq$ 20 dBZ',
    'convection_area':' Area $\geq$ 40 dBZ',
    'intense_area':' Area $\geq$ 50 dBZ',
    'duration':' Storm Duration',
    'mean_intensity':' Intensity',
    'max_intensity':' Maximum Intensity',
    'speed':' Speed',
}

y_units={
    'area':'km$^{2}$',
    'convection_area':'km$^{2}$',
    'intense_area':'km$^{2}$',
    'duration':'hours',
    'mean_intensity':'dBZ',
    'max_intensity':'dBZ',
    'speed':'m s$^{-1}$',
}

season_months={
    'DJF':[12,1,2],
    'MAM':[3,4,5],
    'JJA':[6,7,8],
    'SON':[9,10,11],
    'ALL':np.arange(1,13),
}

letters=['a)','b)','c)','d)','e)','f)','g)','h)','i)','j)']

import matplotlib.lines as mlines

rc45_mark_mwu = mlines.Line2D([], [], color='k', marker='^', linestyle='None',
                          markersize=8)

rc85_mark_mwu = mlines.Line2D([], [], color='k', marker='s', linestyle='None',
                          markersize=8)

rc85_mark_mwu_force = mlines.Line2D([], [], color='k', marker='^', linestyle='None',
                          markersize=8)

rc85_mark_mwu_unforce = mlines.Line2D([], [], color='k', marker='*', linestyle='None',
                          markersize=12)
```

```python
df_name={
    'historical':'historical',
    'end_of_century_8p5':'future_8p5',
    'end_of_century_4p5':'future_4p5'
}

for sim in ['historical', 'end_of_century_8p5']:
    coupled_df = pd.read_csv('/eagle/climate_severe/bwallace_scratch/mcs_pkl_coupled/'+sim+'_grad.csv')
    
    #add category info based on index matching - this will only populate each storm at initiation time!
    data[df_name[sim]].loc[data[df_name[sim]].index, 'category'] = coupled_df['category']
    #this will set the first value for each storm_num as the value for all remaining rows with same storm num in this column
    #need to do this because above line only sets category at initiation time
    data[df_name[sim]]['category'] = data[df_name[sim]].groupby('storm_num')['category'].transform('first')
```

```python
#v_list=['area','convection_area','intense_area','duration','mean_intensity','max_intensity']
v_list=['area','intense_area','duration','speed']

data_store={}

for subset, subset_name in zip(['Forced', 'Unforced', ['Forced','Unforced']], ['Forced', 'Unforced', 'All']):
    #for season in ['ALL','DJF','MAM','JJA','SON']:
    for season in ['ALL','DJF','MAM','JJA','SON']:
        for sim in sims:
            print(subset_name,season,sim)
            data_sub = data[sim][np.isin(data[sim]['category'], subset)]
            for v_i,v in enumerate(v_list):
                for ur_i,ur in enumerate(unique_regions):
                    if season=='ALL':
                        data_store[subset_name+'_'+season+'_'+sim+'_'+v+'_'+ur]=scale_down_data_mean(data_sub,ur,v)
                    else:
                        data_store[subset_name+'_'+season+'_'+sim+'_'+v+'_'+ur]=scale_down_data_mean(
                            data_sub[np.isin(pd.DatetimeIndex(data_sub['datetime']).month, season_months[season])],
                            ur,v)
```
