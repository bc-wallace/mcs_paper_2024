```python
import pickle
import numpy as np
import xarray as xr
import pandas as pd
import wrf

import dask
import dask.array as da
from dask import delayed

import matplotlib.pyplot as plt
from matplotlib import dates as mdates
import matplotlib.colors as mcolors
import matplotlib as mpl
import matplotlib.patheffects as PathEffects
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import matplotlib.patches as patches

import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io import shapereader

import geopandas
import regionmask
from matplotlib.path import Path
import matplotlib.ticker as mticker

from scipy.ndimage.measurements import center_of_mass

from glob import glob

from datetime import datetime, timedelta
import os
```

```python
# Set some projections for our data (Plate Carree)
# and output maps (Lambert Conformal)

# Data projection;
dataproj = ccrs.PlateCarree()

# Plot projection
# The look you want for the view, LambertConformal for mid-latitude view
plotproj = ccrs.LambertConformal(central_longitude=-96.,
                                 central_latitude=40.,
                                 standard_parallels=[30, 60])
```

```python
coarsen_size=20 #make sure this is consistent with the coarsen_size specified in calculate_mcs_anomaly

wrf_native_coords=xr.open_dataset('../data/geog_info.nc')
wrf_lat=wrf_native_coords.XLAT_M[0]
wrf_lon=wrf_native_coords.XLONG_M[0]

wrf_lat_coarse=wrf_lat.coarsen(south_north=coarsen_size,boundary='trim').mean()\
.coarsen(west_east=coarsen_size,boundary='trim').mean()
wrf_lon_coarse=wrf_lon.coarsen(south_north=coarsen_size,boundary='trim').mean()\
.coarsen(west_east=coarsen_size,boundary='trim').mean()

# request data for use by geopandas
resolution = '110m'
category = 'cultural'
name = 'admin_0_countries'

shpfilename = shapereader.natural_earth(resolution, category, name)
df = geopandas.read_file(shpfilename)
df=df[df.TYPE=='Country']
USA=df[df.SOV_A3=='US1']

shpfilename_lakes = shapereader.natural_earth('110m', 'physical', 'lakes')
lakes_df = geopandas.read_file(shpfilename_lakes)

lakes_df=lakes_df[lakes_df.name_alt=='Great Lakes']#.geometry

#glakes_mask=regionmask.mask_geopandas(lakes_df.geometry,wrf_lon,wrf_lat)
#usa_mask=regionmask.mask_geopandas(USA.geometry,wrf_lon,wrf_lat)
```

```python
letters=['a.','b.','c.','d.','e.','f.','g.','h.','i.','k.','l.','m.','n.','o.','p.','q.','r.','s.','t.','u.','v.']
```

```python
allpaths_ds={}
ivt_ds = {}

for sim in ['historical','end_of_century_8p5']:
    for season in ['MAM']:
        allpaths_ds[sim+'_'+season+'_mean']=xr.open_dataset(
        '/eagle/climate_severe/bwallace_scratch/ISO_AVG/'+sim+'/composite_mean/'\
                                                     +season+'_mean.nc')

        ivt_ds[sim+'_'+season+'_mean']=xr.open_dataset(
        '/eagle/climate_severe/bwallace_scratch/ISO_AVG/'+sim+'/composite_mean/'\
                                                     +'IVT_'+season+'_mean.nc')
```

```python
with open('/eagle/climate_severe/bwallace_scratch/paper_repo/iso_avg.pkl', 'rb') as f:
    # Load the dictionary from the pickle file
    allpaths_ds = pickle.load(f)

with open('/eagle/climate_severe/bwallace_scratch/paper_repo/ivt_avg.pkl', 'rb') as f:
    # Load the dictionary from the pickle file
    ivt_ds = pickle.load(f)
```

```python
pres_arr=xr.open_dataset('../data/vert_coords.nc')
```

```python
new_colors=plt.get_cmap('seismic')(np.linspace(0,1,256))
new_colors[120:140]=[1,1,1,1]

seis_cmap=mcolors.ListedColormap(new_colors).resampled(20)
seis_cmap

mean_state_cmap=\
mcolors.ListedColormap(plt.get_cmap('jet')(np.linspace(0.2,0.8,256))).resampled(20)

diff_state_cmap=plt.get_cmap('BrBG').resampled(16)
```

```python
height_ranges={
    90000:{
        'min':975,
        'max':1100,
        'diffmin':-25,
        'diffmax':25,
        'diffrange':10,
    },
    85000:{
        'min':1450,
        'max':1575,
        'diffmin':-50,
        'diffmax':50,
        'diffrange':20,
    },
    70000:{
        'min':3000,
        'max':3200,
        'diffmin':-150,
        'diffmax':150,
        'diffrange':40,
    },
    50000:{
        'min':5600,
        'max':5950,
        'diffmin':-300,
        'diffmax':300,
        'diffrange':120,
    }
}
```

```python
sim='historical'
ssn='MAM'
i=0
xrng,yrng=5, 5


fig=plt.figure(figsize=(13.5,8))
gs=fig.add_gridspec(2,3)

plotnum=0
for i in range(2):
    for j,sel_hgt in enumerate([90000,85000,70000]):
        ax=fig.add_subplot(gs[i,j],projection=plotproj)
        ax.set_extent([-112., -75, 24, 51],ccrs.PlateCarree())
        #ax.add_feature(cfeature.NaturalEarthFeature('physical', 'land', '50m',facecolor=None))
        #ax.add_feature(cfeature.NaturalEarthFeature('physical', 'ocean', '50m',facecolor=None))
        ax.add_feature(cfeature.COASTLINE,linewidth=0.5,zorder=6)
        ax.add_feature(cfeature.STATES,linewidth=0.5,zorder=6)

        ivt_mag_hist = np.sqrt(ivt_ds['historical_'+ssn+'_mean']['IVT_U']**2 + ivt_ds['historical_'+ssn+'_mean']['IVT_V']**2).mean('water_year')
        ivt_mag_eoc = np.sqrt(ivt_ds['end_of_century_8p5_'+ssn+'_mean']['IVT_U']**2 + ivt_ds['end_of_century_8p5_'+ssn+'_mean']['IVT_V']**2).mean('water_year')

        if i==0:
            pmsh = ax.pcolormesh(wrf_lon, wrf_lat,
                          ivt_mag_hist,
                          cmap=mean_state_cmap,
                          vmin = 0, vmax = 100,
                          transform = dataproj, zorder=5)

            contr = ax.contour(wrf_lon_coarse, wrf_lat_coarse,
                               allpaths_ds['historical_'+ssn+'_mean']['GHT_PL'].\
                               mean('water_year').\
                               sel(iso_lev = sel_hgt),
                               np.arange(500, 5000, height_ranges[sel_hgt]['diffrange']),
                               colors='k', linewidths=1.3, linestyles='-',
                               transform=dataproj, zorder=6)

            ax.clabel(contr,contr.levels,fontsize=10)

            q = ax.quiver(wrf_lon_coarse.values[::xrng,::yrng],wrf_lat_coarse.values[::xrng,::yrng],
                     allpaths_ds['historical_'+ssn+'_mean']['U_PL'].\
                     mean('water_year').\
                     sel(iso_lev = sel_hgt).values[::xrng,::yrng],
                     allpaths_ds['historical_'+ssn+'_mean']['V_PL'].\
                     mean('water_year').\
                     sel(iso_lev = sel_hgt).values[::xrng,::yrng],
                     pivot='middle', scale=100,
                     transform=dataproj,zorder=5)
            
            ax.set_title(str(int(sel_hgt/100))+' hPa',fontsize=15)
            ax.title.set_position([0.25, 1])  # Shift title upward to be aligned with the left center
            ax.title.set_ha('left')  # Align title text to the left

        else:
            pmsh = ax.pcolormesh(wrf_lon, wrf_lat,
                          ivt_mag_eoc - ivt_mag_hist,
                          cmap=diff_state_cmap,
                          vmin = -80, vmax = 80,
                          transform = dataproj, zorder=5)

            contr=ax.contour(wrf_lon_coarse,wrf_lat_coarse,
                       allpaths_ds['end_of_century_8p5_'+ssn+'_mean']['GHT_PL'].\
                       mean('water_year').\
                       sel(iso_lev = sel_hgt)-
                       allpaths_ds['historical_'+ssn+'_mean']['GHT_PL'].\
                       mean('water_year').\
                       sel(iso_lev = sel_hgt),
                       np.arange(height_ranges[sel_hgt]['diffmin'],height_ranges[sel_hgt]['diffmax']+50,3),colors='k',linewidths=1.3,
                       transform=dataproj,zorder=6)
            ax.clabel(contr,contr.levels,fontsize=10)

            q=ax.quiver(wrf_lon_coarse.values[::xrng,::yrng],wrf_lat_coarse.values[::xrng,::yrng],
                     allpaths_ds['end_of_century_8p5_'+ssn+'_mean']['U_PL'].\
                     mean('water_year').\
                     sel(iso_lev = sel_hgt).values[::xrng,::yrng]-
                     allpaths_ds['historical_'+ssn+'_mean']['U_PL'].\
                     mean('water_year').\
                     sel(iso_lev = sel_hgt).values[::xrng,::yrng],
                     allpaths_ds['end_of_century_8p5_'+ssn+'_mean']['V_PL'].\
                     mean('water_year').\
                     sel(iso_lev = sel_hgt).values[::xrng,::yrng]-
                     allpaths_ds['historical_'+ssn+'_mean']['V_PL'].\
                     mean('water_year').\
                     sel(iso_lev = sel_hgt).values[::xrng,::yrng],
                     scale=20,
                     pivot='middle',transform=dataproj,zorder=5)


        cbar_axs=[]
        cbar_axs.append(ax)

        if i==0:
            t=ax.text(0.03,0.9,letters[plotnum]+' HIST',
                      horizontalalignment='left',
                      size=12,color='k',
                      path_effects=[PathEffects.withStroke(linewidth=4,foreground='white')],
                      transform=ax.transAxes,zorder=10)

            if j==2:
                axins = inset_axes(
                    ax,
                    width="300%",  # width: 5% of parent_bbox width
                    height="5%",  # height: 50%
                    bbox_to_anchor=(1.075, -12.2, 0.015, 13),
                    bbox_transform=ax.transAxes,
                    borderpad=0,
                )
                
                cbar=plt.colorbar(pmsh,cax=axins,orientation='vertical')
                cbar.ax.tick_params(labelsize=12)
                cbar.set_label('Mean Vertically Integrated IVT\n1000 - 700 hPa (kg m$^{-1}$ s$^{-1}$)',fontsize=12)
    
        elif i>0:
            t=ax.text(0.03,0.9,letters[plotnum]+' EoC85 - HIST',
                      horizontalalignment='left',
                      size=12,color='k',
                      path_effects=[PathEffects.withStroke(linewidth=4,foreground='white')],
                      transform=ax.transAxes,zorder=10)

            if j==2:
                axins = inset_axes(
                    ax,
                    width="300%",  # width: 5% of parent_bbox width
                    height="5%",  # height: 50%
                    bbox_to_anchor=(1.075, -12.2, 0.015, 13),
                    bbox_transform=ax.transAxes,
                    borderpad=0,
                )
                cbar=plt.colorbar(pmsh,cax=axins,orientation='vertical', ticks=[-80, -40, 0, 40, 80])
                cbar.ax.tick_params(labelsize=12)
                cbar.set_label('$\Delta$ Mean Vertically Integrated IVT\n1000 - 700 hPa (kg m$^{-1}$ s$^{-1}$)',fontsize=12)


        t.set_bbox(dict(facecolor='white',alpha=0.8,edgecolor='black'))

        plotnum+=1

        ll=ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,linestyle=':',lw=1.2,zorder=8)
        ll.top_labels=False
        ll.right_labels=False
        if j>0:
            ll.left_labels = False
        if i<1:
            ll.bottom_labels=False
        ll.xlocator = mticker.FixedLocator([-120,-110,-100,-90,-80,-70])
        ll.rotate_labels=True
        ll.xlabel_style = {'rotation':45,  'size':9}
        ll.ylabel_style = {'rotation':15, 'size':9}


        if i==0:
            qk=ax.quiverkey(q,X=0.65 ,Y=1.05,U=5,label='5 m$^{-1}$ s$^{-1}$', transform=dataproj, fontproperties={'weight':'bold','size':13},labelpos = 'E')
        else:
            qk=ax.quiverkey(q,X=0.65,Y=1.03,U=2,label='2 m$^{-1}$ s$^{-1}$', transform=dataproj, fontproperties={'weight':'bold','size':13},labelpos = 'E')



plt.subplots_adjust(wspace=0.04, hspace=-0.02)

plt.show()
plt.close()
```
